## Building our model

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV as gscv
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, ExtraTreesClassifier
from sklearn.ensemble import VotingClassifier
from scipy.stats import expon, uniform
```

```python
df.shape
```

Dropping both "Unnamed: 32" and "Id" columns, to build our model

```python
df = df.drop(labels="Unnamed: 32", axis=1)
df = df.drop(labels="id", axis=1)
```

```python
df.shape
```

```python
df.head()
```

```python
X, y = df.drop('diagnosis', axis=1), df[['diagnosis']]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=43, stratify=y)
```

```python
model_dict = {}
```

```python
# Naive Bayes
classifer = GaussianNB()
predictor = classifer.fit(X_train, y_train)
y_pred = predictor.predict(X_val)
accuracy_naive_bayes = accuracy_score(y_val, y_pred)
model_dict['naive_bayes'] = accuracy_naive_bayes
print(accuracy_naive_bayes)
```

```python
# Logistic Regression
classifier2 = LogisticRegression(random_state=42)
predictor = classifier2.fit(X_train, y_train)
y_pred = predictor.predict(X_val)
accuracy_log_reg = accuracy_score(y_val, y_pred)
model_dict['lg'] = accuracy_log_reg
print(accuracy_log_reg)
```

```python
# Support Vector Classifier
classifier3 = SVC(random_state=42)
predictor_svc = classifier3.fit(X_train, y_train)
y_pred = predictor_svc.predict(X_val)
accuracy_svc = accuracy_score(y_val, y_pred)
model_dict['SVM'] = accuracy_svc
print(accuracy_svc)
```

```python
# Kneighbors
from sklearn.neighbors import KNeighborsClassifier
classifier4 = KNeighborsClassifier()
predictor = classifier4.fit(X_train, y_train)
y_pred = predictor.predict(X_val)
accuracy_knn = accuracy_score(y_val, y_pred)
model_dict['Knn'] = accuracy_knn
print(accuracy_knn)
```

```python
classifier10 = DecisionTreeClassifier(max_depth=3, min_samples_leaf=5, criterion='entropy', min_samples_split=5,
                                       splitter='random', random_state=1)

predictor = classifier10.fit(X_train, y_train)
y_pred = predictor.predict(X_val)
accuracy_dt = accuracy_score(y_val, y_pred)
model_dict['dt'] = accuracy_dt
print(accuracy_dt)
```

```python
#Random Forest Classifier
classifier6 = RandomForestClassifier(random_state=42)
predictor = classifier6.fit(X_train, y_train)
y_pred = predictor.predict(X_val)
accuracy_rfc = accuracy_score(y_val, y_pred)
model_dict['rfc'] = accuracy_rfc
print(accuracy_rfc)
```

```python
classifier7 = GradientBoostingClassifier(random_state=42)
predictor_gbc = classifier7.fit(X_train, y_train)
y_pred = predictor_gbc.predict(X_val)
accuracy_gbc = accuracy_score(y_val, y_pred)
model_dict['gbc'] = accuracy_gbc
print(accuracy_gbc)
```

```python
#xboost
classifier8 = XGBClassifier(random_state=42, eval_metric='logloss')
predictor_xgb = classifier8.fit(X_train, y_train)
y_pred = predictor_xgb.predict(X_val)
accuracy_xgb = accuracy_score(y_val, y_pred)
model_dict['xgb'] = accuracy_xgb
print(accuracy_xgb)
```

```python
#adboost
dtc=DecisionTreeClassifier(criterion='entropy', random_state=42)
classifier = AdaBoostClassifier(dtc,random_state=42)
predictor = classifier.fit(X_train, y_train)
y_pred = predictor.predict(X_val)
accuracy_ada = accuracy_score(y_val, y_pred)
model_dict['adab'] = accuracy_ada
print(accuracy_ada)
```

```python
model_dict
```

```python
model_accuracies_df = pd.DataFrame(columns=['Model', 'Accuracy'])
model_accuracies_df['Model'] = model_dict.keys()
model_accuracies_df['Accuracy'] = model_dict.values()
```

```python
df1 = model_accuracies_df.sort_values(by = "Accuracy", ascending=False)
```

```python
plt.figure(figsize=(18,6))
sns.barplot(df1, x = 'Model', y = 'Accuracy', palette = 'Oranges_r')
plt.title("Model Vs Accuracy for Cancer")
```
